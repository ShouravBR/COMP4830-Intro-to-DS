### Question 2
Write a python code to draw the given line equation at the below:
f(x)= x^7 - 8x^3+ 5x  between -2, 1. 
Use annotating text property to show the maximum of this function.


Your Answer:

Creating a plotting function that also puts arrow at the max y value and returns those (x,y) tuple to reuse code later

```python
def plot_func(f, lim, xlabel='x', ylabel='x', num=1000):
    x = np.linspace(lim[0], lim[1], num = num)
    y = np.array([f(i) for i in x])

    i = np.argmax(y)

    plt.plot(x,y)
    plt.annotate(f'Maximum at ({x[i]:.3f}, {y[i]:.3f})', (x[i], y[i]), (0, -200), textcoords='offset pixels', arrowprops={'width':1})

    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    return x[i], y[i]
```

```python
x_max, y_max = plot_func(lambda x: x**7 - 8*(x**3) + 5*x, [-2,1], ylabel='f(x)')
print(f'maximum y =  {y_max:.3f} for x = {x_max:.3f}')
```
